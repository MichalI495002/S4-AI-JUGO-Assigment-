# Exploratory Data Analysis (EDA) of Maarheeze-Eindhoven Dataset

This notebook presents an exploratory data analysis of a dataset related to events between Maarheeze and Eindhoven. The analysis includes basic data structure examination, summary statistics, missing value analysis, unique value exploration, and data visualization.


```python
import pandas as pd

# Load the dataset
file_path = 'Eindhoven-Arnhem.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

## Data Structure

First, let's examine the structure of the dataset, including the number of rows, columns, and the data types of each column.


```python
# Checking the structure of the dataset
data_structure = {
    "Number of Rows": data.shape[0],
    "Number of Columns": data.shape[1],
    "Column Names": data.columns.tolist(),
    "Data Types": data.dtypes
}

data_structure
```

## Summary Statistics and Missing Values

Next, we generate basic summary statistics for the numerical columns and check for any missing values in the dataset.


```python
# Generating summary statistics for numerical columns
numerical_summary = data.describe()

# Checking for missing values
missing_values = data.isnull().sum()

numerical_summary, missing_values[missing_values > 0]  # Only showing columns with missing values
```

## Unique Values in Categorical Columns

Understanding the diversity in categorical columns is essential. We will explore the unique values in some key categorical columns.


```python
# Exploring unique values in certain categorical columns
unique_values = {
    "Trajectory From": data['Trajectory From'].unique(),
    "Trajectory To": data['Trajectory To'].unique(),
    "Cause Ground Detail": data['Cause Ground Detail'].unique(),

    "Route": data['Route'].unique()
}

unique_values
```

## Data Visualization

Visualizations will help us better understand the distribution and relationship of data in the dataset. We will create plots to visualize the distribution of events over months, frequency of different causes, and severity distribution.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

```

```python
data['File End Date'] = pd.to_datetime(data['File End Date'], format='%Y-%m-%d')

# Extract the month and create a new column 'Month'
data['Month'] = data['File End Date'].dt.month


# Distribution of events over months
sns.countplot(x='Month', data=data, palette='viridis')
plt.title('Distribution of Events Over Months')

# Creating visualizations
plt.figure(figsize=(15, 5))


plt.tight_layout()
plt.show()


```

#### Distribution of Events Over Months

The bar chart illustrates the number of events that occurred in each month throughout a given year. The vertical axis represents the count of events, and the horizontal axis represents the months, labeled 1 through 12.

From the chart, we can observe the following:
- The number of events gradually increases from the beginning of the year, reaching a peak in June (month 6).
- After this peak, there is a significant drop in the number of events in July (month 7), which represents the lowest count across all months.
- Following the dip in July, the frequency of events begins to climb again, with another peak occurring in October (month 10).
- As the year draws to a close, there is a notable decrease in the number of events, with December (month 12) showing fewer events compared to the mid-year highs.


```python
# Frequency of different causes (top 10)
top_causes = data['Cause Ground Detail'].value_counts().head(10)
sns.barplot(y=top_causes.index, x=top_causes.values, palette='rocket')
plt.title('Top 10 Causes of Events')

# Creating visualizations
plt.figure(figsize=(15, 5))


plt.tight_layout()
plt.show()
```

#### Top 10 Causes of Events

The horizontal bar chart outlines the most frequent causes of events within a dataset. The horizontal axis quantifies the frequency of each cause, while the vertical axis lists the causes.

Observations from the chart include:
- The leading cause of events is 'Rush hour traffic jam (no cause reported)', suggesting that the majority of events are related to congestion during peak travel times without a specific cause being reported.
- 'Traffic jam outside rush hour (no cause reported)' ranks as the second most common cause, indicating that traffic-related issues also occur significantly outside of the typical rush hour periods.
- An interesting entry is 'Rush hour traffic jam (no cause reported) with work found in Spin', which may refer to congestion due to roadwork, possibly in a location or with a term specific to the dataset.
- Vehicle-related issues such as 'Defective truck(s)' and 'Defective vehicle(s)' appear as notable causes, as well as 'Accident(s)' and 'Ongeval (met opruim/berging)' which translates to 'Accident (with cleanup/towing)'.
- The chart also includes 'Elevated works' and 'Spedoeraphite', which might be specific events or incorrectly transcribed terms from the dataset that need further clarification.



## Time-Series Analysis

We will conduct a time-series analysis to observe trends and patterns over time. This includes analyzing the distribution of events across different months, time of the day, and the duration of these events.


```python
# Converting date and time columns to datetime format and setting new datetime columns
data['File Start Date'] = pd.to_datetime(data['File Start Date'])
data['File End Date'] = pd.to_datetime(data['File End Date'])
data['Start DateTime'] = pd.to_datetime(data['File Start Date'].astype(str) + ' ' + data['File Start Time'])
data['End DateTime'] = pd.to_datetime(data['File End Date'].astype(str) + ' ' + data['File End Time'])

# Creating time-series visualizations
plt.figure(figsize=(10, 6))
```

```python
data.groupby(data['Start DateTime'].dt.to_period("M")).size().plot(kind='bar')
plt.title('Monthly Event Frequency')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)

# Apply tight layout to adjust the spacing
plt.tight_layout()

# Display the plot
plt.show()
```

#### Monthly Event Frequency Analysis

The bar chart displays the number of events recorded each month from December 2022 to October 2023. The x-axis represents the months of the year, labeled from December of the previous year (2022-12) to October of the current year (2023-10), and the y-axis indicates the count of events that occurred in each month.

Observations from the chart:

- The year begins with a moderate number of events in December (2022-12), which slightly decreases in January (2023-01) before it starts to increase from February (2023-02) onwards.
- A significant peak is observed in May (2023-05), which could suggest a seasonal influence or a period of heightened activity that warrants further investigation.
- The highest frequency of events is recorded in June (2023-06), indicating this month has the most events.
- Following the peak in June, there is a noticeable decline in July (2023-07). Event counts rise again in August (2023-08) and decrease slightly in September (2023-09), before increasing once more in October (2023-10).




```python
# Time of day patterns for event occurrences
data['Start DateTime'].dt.hour.value_counts().sort_index().plot(kind='bar')
plt.title('Time of Day for Event Occurrences')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Events')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### Time of Day for Event Occurrences Analysis

This bar chart represents the distribution of events across different hours of the day. The x-axis denotes the hour of the day in 24-hour format, and the y-axis indicates the number of events occurring at each hour.

Observations from the chart include:

- There is a significant peak at 16:00 (4 PM), which suggests a surge in events during the late afternoon. This could correlate with increased traffic due to the end of the typical workday or school day, leading to more events.
- Another notable increase occurs at 18:00 (6 PM), which is likely associated with the evening rush hour as people commute home.
- The chart shows a steady occurrence of events from the morning through the afternoon (8:00 AM to 3:00 PM), with a slight increase around midday at 11:00 (11 AM).
- After the evening peak, the number of events significantly declines, with much lower counts into the late evening, reaching the lowest levels after 20:00 (8 PM).




```python
# Duration of events over time
data['Event Duration'] = (data['End DateTime'] - data['Start DateTime']).dt.total_seconds() / 60  # Duration in minutes
data.plot(x='Start DateTime', y='Event Duration', kind='line', ax=plt.gca())
plt.title('Duration of Events Over Time')
plt.xlabel('Date')
plt.ylabel('Duration (Minutes)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

#### Duration of Events Over Time Analysis

The line chart depicts the duration of events over the period from January 2023 to November 2023. The x-axis represents the date, and the y-axis represents the duration of events in minutes.

Key observations from the chart:

- Event durations show considerable variability, with most events having shorter durations interspersed with occasional spikes that represent longer-lasting events.
- The occurrence of longer events seems sporadic, without a clear seasonal or monthly pattern, suggesting that these may be associated with specific incidents or extraordinary circumstances.
- There is no discernible trend in event durations over the observed period, indicating that the factors influencing event durations are varied and not necessarily linked to the time of year.
- The chart shows that events occur consistently throughout the year, with no significant periods of inactivity or downtime, as evidenced by the frequent data points across all months.



## Correlation and Categorical Analysis

Now, we'll conduct a correlation analysis among numerical variables and explore the relationships between categorical variables and other key metrics like event severity and duration.


```python
# Correlation Analysis among numerical variables
numerical_data = data[['File Severity', 'Average Length', 'Hectometer Head', 'Hectometer Tail', 'Event Duration']]

# Calculating correlation
corr = numerical_data.corr()

# Creating a heatmap for correlation
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()


```

## Correlation Heatmap Analysis

The correlation heatmap provides a visual summary of the relationships between different numerical variables. Each cell in the heatmap shows the correlation coefficient between two variables, with the scale on the right indicating the strength of the relationship, ranging from -1 (perfect negative correlation) to 1 (perfect positive correlation).

### Key Observations:

- **File Severity and Event Duration**: There is a very high positive correlation (0.95) between 'File Severity' and 'Event Duration', suggesting that events with greater severity tend to have longer durations.

- **Average Length and File Severity**: A notable positive correlation (0.70) between 'Average Length' and 'File Severity' suggests that longer events are typically more severe.

- **Hectometer Head and Hectometer Tail**: There is a perfect positive correlation (1.00) between 'Hectometer Head' and 'Hectometer Tail', which is expected as they likely represent two ends of the same location measure.

- **Other Correlations**: Other variables such as 'Hectometer Head' with 'File Severity' and 'Hectometer Tail' with 'File Severity' show very weak correlations, indicating no significant linear relationship between the location markers and the severity of an event.

- **Event Duration and Average Length**: 'Event Duration' also shows a strong positive correlation with 'Average Length' (0.67), reinforcing the idea that events that last longer tend to be more severe or cover larger areas.




Now we will go more in-depth and look how different 'Cause Ground Detail' categories impact the 'File Severity' and 'Event Duration'. 

```python
# In-depth analysis of categorical variables
# Analyzing the impact of 'Cause Ground Detail' on 'File Severity' and 'Event Duration'
cause_severity = data.groupby('Cause Ground Detail')['File Severity'].mean().sort_values(ascending=False)
cause_duration = data.groupby('Cause Ground Detail')['Event Duration'].mean().sort_values(ascending=False)

plt.figure(figsize=(15, 5))

# Impact on File Severity
cause_severity.head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Causes by Average Severity')
plt.xlabel('Average Severity')
plt.ylabel('Cause')

plt.tight_layout()
plt.show()

```

#### Analysis of Top 10 Causes by Average Severity

The horizontal bar chart lists the top 10 causes of events, ranked by their average severity. The severity is represented on the horizontal axis, scaled to show severity values, which appears to be in the range of hundreds of thousands (as indicated by the `1e6` at the far end of the scale, which denotes scientific notation for one million). Each bar's length represents the average severity score for each cause, with the causes listed on the vertical axis.

Key observations from the chart:

- The most severe cause, by a significant margin, is 'Ongeval (met spoedreparatie)', which translates to 'Accident (with emergency repair)', suggesting that these incidents have the highest average severity.
- 'File buiten spits zonder oorzaak (met werkzaamheden)' translates to 'Traffic jam outside rush hour without cause (with roadwork)', ranking high in severity, indicating that disruptions due to roadwork significantly impact event severity.
- The cause 'Defecte vrachtwagen (in een spitsfile)' or 'Defective truck (in a traffic jam)' also appears as one of the top causes, implying that truck malfunctions during peak traffic contribute notably to the severity of events.
- Other notable causes include various scenarios involving traffic jams and accidents, such as 'Rush hour traffic jam (with accident)' and 'Accident (in a rush hour traffic jam)', which are indicative of the compounded impact these situations have on event severity.
- The least severe causes among the top ten include 'Defecte vrachtwagen (met opruim/berging)' or 'Defective truck (with cleanup/towing)', which, while still significant, are less severe on average compared to the other leading causes.



```python
# Impact on Event Duration
cause_duration.head(10).plot(kind='barh', color='salmon')
plt.title('Top 10 Causes by Average Event Duration')
plt.xlabel('Average Duration (Minutes)')
plt.ylabel('Cause')

plt.tight_layout()
plt.show()
```

#### Analysis of Top 10 Causes by Average Event Duration

This horizontal bar chart depicts the average duration, in minutes, of events attributed to various causes. The x-axis quantifies the average duration while the y-axis lists the causes of events.

Observations from the chart include:

- The leading cause of events in terms of duration is 'File buiten spits zonder oorzaak (met werkzaamheden)', which translates to 'Traffic jam outside rush hour without cause (with roadwork)'. These events have the longest average duration, significantly outstripping other causes.
- 'Ongeval (met spoedreparatie)', or 'Accident (with emergency repair)', also ranks highly, indicating that such incidents result in prolonged event durations, likely due to the urgency and complexity of the repairs needed.
- The event 'Spitsfile (met ongeval vrachtwagen met opruim/berging)', or 'Rush hour traffic jam (with truck accident with cleanup/towing)', represents another significant cause of extended durations, highlighting the impact of truck accidents during peak traffic times.
- Notably, causes related to 'Rush hour traffic jam' with various complications, such as accidents or the presence of defective vehicles, are prominent contributors to longer event durations.
- The cause with the shortest average duration among the top 10 is 'Defecte vrachtwagen (in een spitsfile)', or 'Defective truck (in a traffic jam)', which suggests that while these events are frequent, they are typically resolved more quickly than other types of incidents.
