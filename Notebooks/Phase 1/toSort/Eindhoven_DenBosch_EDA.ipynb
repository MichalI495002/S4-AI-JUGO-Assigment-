# Exploratory Data Analysis (EDA) of Maarheeze-Eindhoven Dataset

This notebook presents an exploratory data analysis of a dataset related to events between Maarheeze and Eindhoven. The analysis includes basic data structure examination, summary statistics, missing value analysis, unique value exploration, and data visualization.


```python
import pandas as pd

# Load the dataset
file_path = 'Eindhoven-DenBosch.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

## Data Structure

First, let's examine the structure of the dataset, including the number of rows, columns, and the data types of each column.


```python
# Checking the structure of the dataset
data_structure = {
    "Number of Rows": data.shape[0],
    "Number of Columns": data.shape[1],
    "Column Names": data.columns.tolist(),
    "Data Types": data.dtypes
}

data_structure
```

## Summary Statistics and Missing Values

Next, we generate basic summary statistics for the numerical columns and check for any missing values in the dataset.


```python
# Generating summary statistics for numerical columns
numerical_summary = data.describe()

# Checking for missing values
missing_values = data.isnull().sum()

numerical_summary, missing_values[missing_values > 0]  # Only showing columns with missing values
```

## Unique Values in Categorical Columns

Understanding the diversity in categorical columns is essential. We will explore the unique values in some key categorical columns.


```python
# Exploring unique values in certain categorical columns
unique_values = {
    "Trajectory From": data['Trajectory From'].unique(),
    "Trajectory To": data['Trajectory To'].unique(),
    "Cause Ground Detail": data['Cause Ground Detail'].unique(),

    "Route": data['Route'].unique()
}

unique_values
```

## Data Visualization

Visualizations will help us better understand the distribution and relationship of data in the dataset. We will create plots to visualize the distribution of events over months, frequency of different causes, and severity distribution.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

```

```python
data['File End Date'] = pd.to_datetime(data['File End Date'], format='%Y-%m-%d')

# Extract the month and create a new column 'Month'
data['Month'] = data['File End Date'].dt.month


# Distribution of events over months
sns.countplot(x='Month', data=data, palette='viridis')
plt.title('Distribution of Events Over Months')

# Creating visualizations
plt.figure(figsize=(15, 5))


plt.tight_layout()
plt.show()


```

The bar chart illustrates the number of events that occurred in each month throughout a given year. The vertical axis represents the count of events, and the horizontal axis represents the months, labeled 1 through 12.

From the chart, we can observe the following:
- The number of events gradually increases from the beginning of the year, with a noticeable peak in June (month 6).
- There is a significant drop in events after June, with the lowest counts occurring in July (month 7).
- The event frequency begins to rise again after July, peaking once more in October (month 10).
- The count decreases towards the end of the year, with December (month 12) showing a lower number of events than the mid-year peak.



```python
# Frequency of different causes (top 10)
top_causes = data['Cause Ground Detail'].value_counts().head(10)
sns.barplot(y=top_causes.index, x=top_causes.values, palette='rocket')
plt.title('Top 10 Causes of Events')

# Creating visualizations
plt.figure(figsize=(15, 5))


plt.tight_layout()
plt.show()
```

The horizontal bar chart outlines the most frequent causes of events within a dataset. The horizontal axis quantifies the frequency of each cause, while the vertical axis lists the causes.

Observations from the chart include:
- The leading cause of events is 'Rush hour traffic jam (no cause reported)', suggesting that the majority of events are related to congestion during peak travel times without a specific cause being reported.
- The second most common cause is 'Traffic jam outside rush hour (no cause reported)', which indicates that significant traffic-related issues occur outside of the typical rush hour times.
- The third on the list is 'Rush hour traffic jam (no cause reported) with work found in Spin', highlighting specific instances where traffic congestion is compounded by work events.
- Further notable causes include vehicle-related issues such as 'Defective truck(s)' and 'Defective vehicle(s)', as well as 'Accident(s)' and 'Ongeval (met opruim/berging)', which translates to 'Accident (with cleanup/towing)'.
- The chart also draws attention to less frequent but impactful causes such as 'Ongeval op aansluitende weg (met spitsfile)' or 'Accident on connecting road (with traffic jam)', emphasizing the range of factors contributing to events.



## Time-Series Analysis

We will conduct a time-series analysis to observe trends and patterns over time. This includes analyzing the distribution of events across different months, time of the day, and the duration of these events.


```python
# Converting date and time columns to datetime format and setting new datetime columns
data['File Start Date'] = pd.to_datetime(data['File Start Date'])
data['File End Date'] = pd.to_datetime(data['File End Date'])
data['Start DateTime'] = pd.to_datetime(data['File Start Date'].astype(str) + ' ' + data['File Start Time'])
data['End DateTime'] = pd.to_datetime(data['File End Date'].astype(str) + ' ' + data['File End Time'])

# Creating time-series visualizations
plt.figure(figsize=(10, 6))
```

```python
data.groupby(data['Start DateTime'].dt.to_period("M")).size().plot(kind='bar')
plt.title('Monthly Event Frequency')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)

# Apply tight layout to adjust the spacing
plt.tight_layout()

# Display the plot
plt.show()
```

#### Monthly Event Frequency Analysis

The bar chart displays the number of events recorded each month from December 2022 to October 2023. The x-axis represents the months of the year, labeled from December of the previous year (2022-12) to October of the current year (2023-10), and the y-axis indicates the count of events that occurred in each month.

Observations from the chart:

- The year begins with a moderate number of events in December (2022-12), which dips in January (2023-01) and then gradually increases through February (2023-02) to April (2023-04).
- There is a marked increase in events in May (2023-05), suggesting a possible seasonal influence or a period of heightened activity.
- The most significant peak occurs in June (2023-06), indicating the highest frequency of events during this month.
- After this peak, there's a sharp decline in July (2023-07) followed by a steady increase through to October (2023-10), excluding a minor decrease in September (2023-09).



```python
# Time of day patterns for event occurrences
data['Start DateTime'].dt.hour.value_counts().sort_index().plot(kind='bar')
plt.title('Time of Day for Event Occurrences')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Events')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#### Time of Day for Event Occurrences Analysis

This bar chart represents the distribution of events across different hours of the day. The x-axis denotes the hour of the day in 24-hour format, and the y-axis indicates the number of events occurring at each hour.

Observations from the chart include:

- The chart shows a significant peak at 16:00 (4 PM), which shows a surge in events during the late afternoon, possibly correlating with the end of the typical workday or school day.
- The next notable increase occurs at 18:00 (6 PM), likely associated with evening rush hour as people commute home.
- There is a steady presence of events during the hours of 8:00 (8 AM) to 15:00 (3 PM), with a slight increase around 11:00 (11 AM).
- The number of events significantly tapers off after the peak at 18:00 (6 PM), with much lower counts into the evening hours, reaching the lowest levels after 20:00 (8 PM).




```python
# Duration of events over time
data['Event Duration'] = (data['End DateTime'] - data['Start DateTime']).dt.total_seconds() / 60  # Duration in minutes
data.plot(x='Start DateTime', y='Event Duration', kind='line', ax=plt.gca())
plt.title('Duration of Events Over Time')
plt.xlabel('Date')
plt.ylabel('Duration (Minutes)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

#### Duration of Events Over Time Analysis

The line chart depicts the duration of events over the period from January 2023 to November 2023. The x-axis represents the date, and the y-axis represents the duration of events in minutes.

Key observations from the chart:

- Event durations exhibit considerable variability, with most events having shorter durations interspersed with occasional spikes representing longer-lasting events.
- The occurrence of longer events is sporadic and does not follow a clear seasonal or monthly pattern, suggesting they may be associated with specific incidents or extraordinary circumstances.
- No discernible trend of increasing or decreasing event duration over time is observed, indicating that the factors influencing event durations are diverse and may not be linked to any particular time of the year.
- The presence of numerous data points throughout the year suggests a consistent occurrence of events with no significant periods of inactivity or downtime.


## Correlation and Categorical Analysis

Now, we'll conduct a correlation analysis among numerical variables and explore the relationships between categorical variables and other key metrics like event severity and duration.


```python
# Correlation Analysis among numerical variables
numerical_data = data[['File Severity', 'Average Length', 'Hectometer Head', 'Hectometer Tail', 'Event Duration']]

# Calculating correlation
corr = numerical_data.corr()

# Creating a heatmap for correlation
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()


```

## Correlation Heatmap Analysis

The correlation heatmap provides a visual summary of the relationships between different numerical variables. Each cell in the heatmap shows the correlation coefficient between two variables, with the scale on the right indicating the strength of the relationship, ranging from -1 (perfect negative correlation) to 1 (perfect positive correlation).

### Key Observations:

- **File Severity and Event Duration**: There is a very high positive correlation (0.91) between 'File Severity' and 'Event Duration', suggesting that events with greater severity tend to have longer durations.

- **Hectometer Head and Hectometer Tail**: These two variables have a very high positive correlation (0.98), indicating that as the head hectometer marker increases, the tail marker tends to increase correspondingly, which is expected since they are part of the same event location.

- **Average Length**: 'Average Length' has a moderate positive correlation with 'File Severity' (0.53), implying that more severe events are also longer in length.

- **Weak Correlations**: Several variables show weak correlations with others. For example, 'Hectometer Head' and 'File Severity', as well as 'Hectometer Tail' and 'File Severity', have low correlations (around -0.04 and 0.01 respectively), indicating no significant linear relationship between them.


Now we will go more in-depth and look how different 'Cause Ground Detail' categories impact the 'File Severity' and 'Event Duration'. 

```python
# In-depth analysis of categorical variables
# Analyzing the impact of 'Cause Ground Detail' on 'File Severity' and 'Event Duration'
cause_severity = data.groupby('Cause Ground Detail')['File Severity'].mean().sort_values(ascending=False)
cause_duration = data.groupby('Cause Ground Detail')['Event Duration'].mean().sort_values(ascending=False)

plt.figure(figsize=(15, 5))

# Impact on File Severity
cause_severity.head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Causes by Average Severity')
plt.xlabel('Average Severity')
plt.ylabel('Cause')

plt.tight_layout()
plt.show()

```

#### Analysis of Top 10 Causes by Average Severity

The horizontal bar chart lists the top 10 causes of events, ranked by their average severity. The severity is represented on the horizontal axis, scaled to show severity values, which appears to be in the range of hundreds of thousands (as indicated by the `1e6` at the far end of the scale, which denotes scientific notation for one million). Each bar's length represents the average severity score for each cause, with the causes listed on the vertical axis.

Key observations from the chart:

- The most severe cause, by a significant margin, is 'Spitsfile', suggesting that traffic jams during peak hours, especially those involving accidents, are the most severe on average.
- Causes involving accidents, such as 'Ongeval (met opruim/berging)' which translates to 'Accident (with cleanup/towing)', and accidents involving trucks ('Spitsfile met ongeval vrachtauto' - 'Traffic jam with truck accident') also show high average severities, indicating these incidents are particularly impactful.
- The presence of 'Modder op het wegdek' (Mud on the road) as one of the causes indicates that even non-accident-related events can have a significant severity impact.
- The less severe causes among the top ten are related to defective vehicles ('Defecte vrachtwagen' - 'Defective truck') and rush hour traffic jams with or without accidents, which, while still impactful, have lower average severity scores compared to the top-ranking causes.
- It is interesting to note the varied nature of causes, ranging from 'Rush hour traffic jam' to 'Defective vehicle', illustrating the diverse factors that can contribute to the severity of events on the road.


```python
# Impact on Event Duration
cause_duration.head(10).plot(kind='barh', color='salmon')
plt.title('Top 10 Causes by Average Event Duration')
plt.xlabel('Average Duration (Minutes)')
plt.ylabel('Cause')

plt.tight_layout()
plt.show()
```

#### Analysis of Top 10 Causes by Average Event Duration

This horizontal bar chart depicts the average duration, in minutes, of events attributed to various causes. The x-axis quantifies the average duration while the y-axis lists the causes of events.

Observations from the chart include:

- 'Spitsfile' leads the chart, which typically translates to 'traffic jam during rush hour'. Events associated with 'Spitsfile', especially those involving an accident, tend to last significantly longer than those associated with other causes.
- 'Defecte vrachtwagen (met ongeval en opruim/berging)' which translates to 'Defective truck (with accident and cleanup/towing)', and 'Ongeval (met ongeval op aansluitende weg)' meaning 'Accident (with accident on connecting road)', follow as the second and third leading causes for prolonged event durations, respectively.
- Other causes such as 'Ongeval' (Accident), 'Defecte vrachtwagen' (Defective truck), and 'Spitsfile (met ongeval op aansluitende weg)' (Traffic jam with accident on connecting road) are also significant contributors to longer event durations, though to a lesser extent than the top-ranking causes.
- The causes related to defective vehicles and accidents in rush hour seem to be particularly problematic, indicating these events can cause substantial delays.
- The cause with the least average duration among the top 10 is 'Modder op het wegdek' (Mud on the road), shows that these events are resolved more quickly compared to others.
